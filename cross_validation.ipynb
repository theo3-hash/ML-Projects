```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

# Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Data Collection & Processing

```python
heart_data = pd.read_csv('/content/heart.csv')
```

```python
heart_data.head()

```

```python
heart_data.shape
```

```python
heart_data.isnull().sum()
```

# Features & Target splitting

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

# Train & Test data splitting

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Comparing different Model Performance

```python
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()] # list of models
```

```python
def compare_models_train_test():
  for model in models:

    # Training the model
    model.fit(X_train, Y_train)

    # Predicting from the model
    test_data_prediction = model.predict(X_test)
    accuracy = accuracy_score(Y_test, test_data_prediction)
    print('Accuracy score of the ', model, ' = ', accuracy)
```

```python
compare_models_train_test()
```

# Cross Validation

Logistic Regression

```python
cv_score_lr = cross_val_score(LogisticRegression(max_iter=10000), X, Y, cv=5)

print(cv_score_lr)

mean_accuracy_lr = sum(cv_score_lr)/len(cv_score_lr)

mean_accuracy_lr = mean_accuracy_lr*100

mean_accuracy_lr = round(mean_accuracy_lr, 2) # provide 2 deciman

print(mean_accuracy_lr)
```

Support Vector Machine

```python
cv_score_svc = cross_val_score(SVC(kernel='linear'), X, Y, cv=5)

print(cv_score_svc)

mean_accuracy_svc = sum(cv_score_svc)/len(cv_score_svc)

mean_accuracy_svc = mean_accuracy_svc*100

mean_accuracy_svc = round(mean_accuracy_svc, 2) # provide 2 deciman

print(mean_accuracy_svc)
```

K-Nearest Classifier

```python
cv_score_knn = cross_val_score(KNeighborsClassifier(), X, Y, cv=5)

print(cv_score_knn)

mean_accuracy_knn = sum(cv_score_knn)/len(cv_score_knn)

mean_accuracy_knn = mean_accuracy_knn*100

mean_accuracy_knn = round(mean_accuracy_knn, 2) # provide 2 deciman

print(mean_accuracy_knn)
```

Random Forest Classifier

```python
cv_score_rfc = cross_val_score(RandomForestClassifier(), X, Y, cv=5)

print(cv_score_rfc)

mean_accuracy_rfc = sum(cv_score_rfc)/len(cv_score_rfc)

mean_accuracy_rfc = mean_accuracy_rfc*100

mean_accuracy_rfc = round(mean_accuracy_rfc, 2) # provide 2 deciman

print(mean_accuracy_rfc)
```

# Creating a function to compare the models

```python
# List of models
models_list = [LogisticRegression(max_iter=10000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_cross_validation():
  for model in models_list:

    cv_score = cross_val_score(model, X, Y, cv=5)

    mean_accuracy = sum(cv_score)/len(cv_score)

    mean_accuracy = mean_accuracy*100

    mean_accuracy = round(mean_accuracy, 2) # for 2 decimal number

    print('Cross Validation accuracies for the ', model, ' = ', cv_score)

    print('Accuracy % of the ', model, ' = ', mean_accuracy)

    print('----------------------------------------------------------------------------------------------------------------')
```

```python
compare_models_cross_validation()
```
