```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
import sklearn.metrics as metrics
```

Data Collection

```python
calories_data = pd.read_csv("E:/Downloads/Calories Brunt Prediction/archive/calories.csv")
exercise_data = pd.read_csv("E:/Downloads/Calories Brunt Prediction/archive/exercise.csv")
```

Combining The Two Dataframes

```python
calories_exercise_data = pd.concat([exercise_data,calories_data['Calories']],axis =1)
```

```python
calories_exercise_data = calories_exercise_data.replace({"Gender":{"male":0,"female":1}})
```

Correlation

```python
correlation = calories_exercise_data.corr()
```

```python
# Constructing a heatmap to understand the correlation between the features
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='Blues', center=0, cbar=True, square=True, annot_kws={'size': 10})
plt.title('Correlation Heatmap')
plt.show()
```

Seperating Features and Targets

```python
X = calories_exercise_data.drop(columns=['Calories'], axis=1)
Y = calories_exercise_data['Calories']
```

```python
print(X)
print(Y)
```

Splitting the data into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

XGBoost Model Training

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

Evaluation

```python
# Preciction on test data
Y_pred = model.predict(X_test)
print(Y_pred)
```

Mean Absolute Error

```python
mae = metrics.mean_absolute_error(Y_test, Y_pred)
print("Mean Absolute Error:", mae)
```
